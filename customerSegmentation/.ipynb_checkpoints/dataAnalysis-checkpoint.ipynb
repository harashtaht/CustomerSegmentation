## Customer Segmentation for Automobile Company

<u>Context</u>

An automobile company has plans to enter new markets with their existing products (P1, P2, P3, P4 and P5). After intensive market research, theyâ€™ve deduced that the behavior of new market is similar to their existing market.

<u>Content</u>

In their existing market, the sales team has classified all customers into 4 segments (A, B, C, D ). Then, they performed segmented outreach and communication for different segment of customers. This strategy has work exceptionally well for them. They plan to use the same strategy on new markets and have identified 2627 new potential customers.

You are required to help the manager to predict the right group of the new customers.

```python
#Function to describe Dataframe

def describeDF(x):
    listItem = []
    for col in x.columns:
        listItem.append([col, x[col].dtype,
                       x[col].isna().sum(),
                       round((x[col].isna().sum()/len(x[col])) *100, 2),
                       x[col].nunique(), list(x[col].unique()[:5])]);

    desc = pd.DataFrame(columns=['dataFeatures', 'dataType', 'null', 'nullPct', 'unique', 'uniqueSample'],
                         data=listItem)
    return desc
```

```python
def plotting(y_, hue_, data_):
    g = sns.catplot(y=y_, hue=hue_, kind="count",
                palette ="Set2", data=data_, 
                height=6, aspect=11.7/9)

    plt.title(f'Distribution of {hue_} in Terms of {y_}')
    plt.xlabel("Frequency")

    total = len(trMod["Segmentation"])
    for ax in g.axes.ravel():
        for p in ax.patches:
            percentage = '{:.1f}%'.format(100 * p.get_width()/total)
            x = p.get_x() + p.get_width() + 0.04
            y = p.get_y() + p.get_height()/2
            ax.annotate(percentage, (x, y))
    plt.grid(True)
```

```python
# Change the value of Spending_Score into number
# Since it is ordinal data, we can use manual Encoder to speed up the model later on

X_2 = X_.copy(deep=True)
X_2['Spending_Score'] = X_['Spending_Score'].apply(lambda val: 2 if val == 'High' else (1 if val == 'Average' else 0))

# Change the value of Gender into number
# Since Male tends to spend more, we label Male as 1 and Female as 0

X_3 = X_2.copy(deep=True)
X_3['Gender'] = X_['Gender'].apply(lambda val: 1 if val == 'Male' else 0)
X_3.head()
```

### 4.b. Preprocessing and Encoding Data

```python
dfMod = dfTr.copy(deep=True)

# Separating dependent and independent variables
X = dfMod.drop('Segmentation', axis=1)
y = dfMod['Segmentation']
```

```python
# Encoding Data

X['Gender'] = X['Gender'].apply(lambda val:1 if val == 'Male' else 0)
X['Ever_Married'] = X['Ever_Married'].apply(lambda val:1 if val == 'Yes' else 0)
X['Graduated'] = X['Graduated'].apply(lambda val:1 if val == 'Yes' else 0)
# X['Spending_Score'] = X['Spending_Score'].apply(lambda val:3 if val == 'High' else (1 if val == 'Low' else 2))
```

```python
# Dividing Range numbers into groups
# The division of the group is based on its distribution in terms of percentiles

# Divide Age into Groups
Age_Bin = []
for i, j in zip(X['Age'], range(len(X['Age']))):
    if (i <= 30):
        Age_Bin.append("17-30")
    elif (30 < i <= 45):
        Age_Bin.append("31-45")
    elif (45 < i <= 60):
        Age_Bin.append("46-60")
    else:
        Age_Bin.append("60+")
X['Age_Bin'] = Age_Bin
X.drop('Age', axis=1, inplace=True)
        
# Divide Work Experience into Groups
X['WorkExpCat'] = pd.cut(X.Work_Experience,bins=[-1,1,7,15],labels=['Low','Medium','High'])
X.drop('Work_Experience', axis=1, inplace=True)

# Divide Family Size into Groups
X['FamSizeCat'] = pd.cut(X.Family_Size,bins=[0,4,6,10],labels=['Small','Medium','Big'])
X.drop('Family_Size', axis=1, inplace=True)
```

```python
# Encoding the data
# For the numerical data that's divided into groups, we use One Hot Encoder for better model performance 
# Also Spending_Score, since it is an ordinal data

# X.rename(columns={'Spending_Score': 'Spending'}, inplace=True)

# Encoding Age
ag = pd.get_dummies(X.Age_Bin, prefix='AgeBin')
X = pd.concat([X, ag], axis = 1).drop('Age_Bin', axis=1)

# Encoding Work Experience
wr = pd.get_dummies(X.WorkExpCat, prefix='WorkExp')
X = pd.concat([X, wr], axis = 1).drop('WorkExpCat', axis=1)

# Encoding Family Size
fm = pd.get_dummies(X.FamSizeCat, prefix='FamSize')
X = pd.concat([X, fm], axis = 1).drop('FamSizeCat', axis=1)

# Encoding Profession
pr = pd.get_dummies(X.Profession, prefix='Profession')
X = pd.concat([X, pr], axis = 1).drop('Profession', axis=1)

# Encoding Var_1
vr = pd.get_dummies(X.Var_1, prefix='Var_1')
X = pd.concat([X, vr], axis = 1).drop('Var_1', axis=1)

# Encoding Spending Score
sp = pd.get_dummies(X.Spending_Score, prefix='Spending')
X = pd.concat([X, sp], axis = 1).drop('Spending_Score', axis=1)


# Since ID will not affect the simulation, it is dropped to avoid confusion
X.drop('ID', axis=1, inplace=True)
X.columns
```

## 5. Machine Learning Modeling

### 5.A. Import Libraries and Base Model Simulation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, accuracy_score, recall_score, auc, log_loss, roc_auc_score, f1_score, matthews_corrcoef
```

```python
def calc_train_error(X_train, y_train, model):
    # return in-sample error for already fit model
    
    predictions = model.predict(X_train)
    predictProba = model.predict_proba(X_train)
    matt = matthews_corrcoef(y_train, predictions)
    f1 = f1_score(y_train, predictions, average='macro')
    classreport = classification_report(y_train, predictions)
    roc_auc = roc_auc_score(y_train, predictProba[:,1])
    accuracy = accuracy_score(y_train, predictions)
    recall = recall_score(y_train, predictions)
    confMatrix = confusion_matrix(y_train, predictions)
    logloss = log_loss(y_train, predictProba)
    return {
        'classreport' : classreport,
        'matthews' : matt,
        'f1' : f1,
        'roc' : roc_auc,
        'accuracy' : accuracy,
        'recall' : recall,
        'confusion' : confMatrix,
        'logloss' : logloss
    }

def calc_validation_error(X_test, y_test, model):
    # return out of sample error for already fit model
    
    predictions = model.predict(X_test)
    predictProba = model.predict_proba(X_test)
    matt = matthews_corrcoef(y_test, predictions)
    f1 = f1_score(y_test, predictions, average='macro')
    classreport = classification_report(y_test, predictions)
    roc_auc = roc_auc_score(y_test, predictProba[:,1])
    accuracy = accuracy_score(y_test, predictions)
    recall = recall_score(y_test, predictions)
    confMatrix = confusion_matrix(y_test, predictions)
    logloss = log_loss(y_test, predictProba)
    return {
        'classreport' : classreport,
        'matthews' : matt,
        'f1' : f1,
        'roc' : roc_auc,
        'accuracy' : accuracy,
        'recall' : recall,
        'confusion' : confMatrix,
        'logloss' : logloss
    }

def calc_metrics(X_train, y_train, X_test, y_test, model):
    # fits model and returns the classification metrics for in sample error and out of sample error
    model.fit(X_train, y_train)
    train_error = calc_train_error(X_train, y_train, model)
    validation_error = calc_validation_error(X_test, y_test, model)
    return train_error, validation_error
```

```python
## Decision Tree
print("*************** Decision Tree *************** \n")

print("\nTrain Data\n")
# Fitting train data into model
baseDT = DecisionTreeClassifier(random_state=27, criterion='gini')
baseDT.fit(X_train, y_train)
print('Accuracy Score: ', baseDT.score(X_train, y_train), "\n")

# Predict the target on the train data 
tr_1_DT = baseDT.predict(X_train)
# pp_tr_DT1 = baseDT.predict_proba(X_train)[:, 1]

# Printing report
print('-------------- Train Data Modeling Report --------------\n')
print(classification_report(y_train, tr_1_DT))
print('-------------- Confusion Matrix --------------\n')
print(confusion_matrix(y_train, tr_1_DT))
print('\n')

print("... Fitting into Test Data\n")
# Simulating on Test Data
print("\nTest Data")
print('Accuracy Score: ', baseDT.score(X_test, y_test), "\n")
predDT = baseDT.predict(X_test)
pp_ts_DT1 = baseDT.predict_proba(X_test)[:,1]
print('-------------- Test Data Modeling Report --------------\n')
print(classification_report(y_test, predDT))
print('-------------- Confusion Matrix --------------\n')
print(confusion_matrix(y_test, predDT))
print('\n')
```
